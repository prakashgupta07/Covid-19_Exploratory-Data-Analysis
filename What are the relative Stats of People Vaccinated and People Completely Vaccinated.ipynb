## [4.10 What are the relative Stats of People Vaccinated and People Completely Vaccinated?](#section4010)

```python
data=Vaccinations.groupby('country')[["date",'people_vaccinated','people_fully_vaccinated']].max()
data_df=pd.DataFrame(data).reset_index()
data_df
```

```python
plt.figure(figsize=(40,10))
fig=sns.lineplot(x='country',y='people_vaccinated',data=data_df,legend = 'brief')
fig=sns.lineplot(x='country',y='people_fully_vaccinated',data=data_df,legend = 'brief')
plt.ylabel("people_vaccinated and people_fully_vaccinated")
plt.title("What are the relative Stats of People Vaccinated and People Completely Vaccinated")
plt.xticks(rotation=90)
plt.show()
```
